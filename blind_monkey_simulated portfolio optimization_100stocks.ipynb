### individual stock analysis

```python

def stock_KPIs(dataframe, start_date, end_date):
    '''
    Explanation
    Daily Returns Calculation: The calculate_individual_returns function calculates returns per stock, leaving NaN where data is missing, so no artificial data is introduced.
    Mean Returns Calculation: Using mean(skipna=True) calculates the mean for each stock based only on existing data, ignoring NaN values.
    Covariance Matrix Calculation: The covariance matrix is calculated using pairwise deletion (min_periods=1), so each covariance value is based only on dates where both stocks have available data.
    '''

    filtered_df = dataframe.loc[start_date:end_date] # option to slice the data frame by date
    filtered_df = filtered_df.where(filtered_df > 0) # Replace zero or negative values with NaN (to handle cases where data might be invalid)
    filtered_df = filtered_df.ffill()  # Forward-fill only for zero or negative values that are transformed to NaN, Note: This still keeps original missing values as NaN


    def calculate_individual_returns(df):
        # Calculate daily log returns without filling NaNs (preserves missing data)
        return df.apply(lambda x: np.log(x / x.shift(1)))

    p_daily_returns = calculate_individual_returns(filtered_df) # still keeps the non-trading periods
    mean_returns = p_daily_returns.mean(skipna=True) * 252  # Annualized mean returns ignoring NaNs
    variance = p_daily_returns.var(skipna=True) * 252 # Annualized variance ignoring NaNs
    volatility = p_daily_returns.std(skipna=True) * np.sqrt(252)  # Annualized standard deviation
    cov_matrix_pairwise = p_daily_returns.cov(min_periods=1) * 252  # Annualized pairwise covariance matrix

    return {
        'data_frame': filtered_df,
        'mean_returns': mean_returns,             # Annualized mean returns
        'variance': variance,
        'volatility': volatility,            # Annualized standard deviation
        'cov_matrix': cov_matrix_pairwise,  # Annualized pairwise covariance matrix
    }
```

```python
# to check - analysis shall use annualized figures
mean_returns = stock_KPIs(stock_close, start_date, end_date)['mean_returns'].rename('mean_returns')
cov_matrix = stock_KPIs(stock_close, start_date, end_date)['cov_matrix']
```

### Functions

```python
def portfolio_return(weights, mean_returns):
  weights_array = np.array(weights)
  mean_returns = np.dot(weights_array,mean_returns)
  return mean_returns
```

```python
def portfolio_volatility(weights, cov_matrix):
  weights_array = np.array(weights)
  variance = np.dot(weights_array.T, np.dot(cov_matrix, weights_array))
  return np.sqrt(variance)
```

```python
def portfolio_variance(weights, cov_matrix):
  weights_array = np.array(weights)
  variance = np.dot(weights_array.T, np.dot(cov_matrix, weights_array))
  return variance
```

```python
def generate_random_weights(num_assets, num_iteration):
  return np.random.dirichlet(np.ones(num_assets), size=num_iteration)
```

```python
def generate_portfolio_data(metrics, cov_matrix, num_portfolios, risk_free_rate=0.05):
    num_assets = len(metrics)  # Number of assets based on the mean_returns length
    mean_returns = metrics['mean_returns'].values
    random_weights = generate_random_weights(num_assets, num_portfolios)  # Generate random weights

    # Prepare an empty list to store portfolio data
    portfolio_data = []

    # Loop through each weight set
    for weight in random_weights:
        port_return = portfolio_return(weight, mean_returns)  # Calculate portfolio return
        port_variance = portfolio_variance(weight, cov_matrix)
        port_volatility = portfolio_volatility(weight, cov_matrix) # Calculate portfolio volatility
        port_sharpe = (port_return - risk_free_rate) / port_volatility

        # Append the results as a dictionary for each portfolio (without the redundant index)
        portfolio_data.append({
            'weights': list(weight),
            'returns': port_return,
            'variance': port_variance,
            'volatility': port_volatility,
            'sharpe ratio': port_sharpe
        })

    # Convert the list of portfolio data into a DataFrame (without duplicate index)
    portfolio_df = pd.DataFrame(portfolio_data)
    return portfolio_df
```

```python
def calculate_betas(max_sharpe_weights, metrics, cov_matrix, ):
    """
    Calculate beta values for a given set of portfolio stocks.

    Parameters:
    - portfolio_stocks: List of selected stock tickers in the maximum Sharpe portfolio.
    - metrics: DataFrame containing returns and volatility data for all stocks.
    - cov_matrix: Full covariance matrix of all stocks.
    - portfolio_weights: Weight vector for the stocks in the maximum Sharpe portfolio.

    Returns:
    - A DataFrame with beta values for each stock in the selected portfolio.
    """
    # Filter the metrics and covariance matrix for the selected stocks
    portfolio_stocks = max_sharpe_weights.index
    portfolio_weights = max_sharpe_weights['weights']
    filtered_metrics = metrics.loc[portfolio_stocks]
    filtered_cov_matrix = cov_matrix.loc[portfolio_stocks, portfolio_stocks]

    # Calculate market portfolio variance based on selected weights
    market_var = portfolio_weights.T @ filtered_cov_matrix.to_numpy() @ portfolio_weights

    # Initialize DataFrame to store beta values
    beta_values = pd.Series(index=portfolio_stocks, dtype=float)

    # Calculate beta for each stock in the selected portfolio
    for i, ticker in enumerate(portfolio_stocks):
        # Create binary vector to isolate the stock
        only_stock = np.zeros(len(portfolio_stocks))
        only_stock[i] = 1  # Set current stock's position to 1

        # Calculate covariance of stock with the market portfolio
        stock_cov_with_market = only_stock.T @ filtered_cov_matrix.to_numpy() @ portfolio_weights
        beta_values[ticker] = stock_cov_with_market / market_var

    # Combine beta values with returns for the output
    result = pd.DataFrame({
        'Return': filtered_metrics['mean_returns'],
        'Beta': beta_values
    })
    return result

```

```python
beta_df = calculate_betas(max_sharpe_weights, metrics, cov_matrix)
beta_df
```

```python
# Extract Beta and Return values for regression
x = beta_df['Beta']
y = beta_df['Return']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Calculate the regression line
regression_line = slope * x + intercept

# Plot the scatter plot and regression line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='black', alpha=0.1, label='Stocks')
plt.plot(x, regression_line, color='orange', linestyle='--', linewidth=1, label=f'Regression Line (y = {slope:.2f}x + {intercept:.2f})')

# Add labels for each stock
for ticker in beta_df.index:
    plt.text(beta_df.loc[ticker, 'Beta'], beta_df.loc[ticker, 'Return'], ticker, color='grey', alpha=0.5, fontsize=9, ha='right')

# Display R-squared value on the plot
plt.text(0.8, 0.2, f'$R^2$ = {r_value**2:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

# Set plot labels and title
plt.xlabel('Beta')
plt.ylabel('Return')
plt.title('Stock Returns vs. Beta with Regression Line')
plt.grid(True)
plt.legend()
plt.show()
```
